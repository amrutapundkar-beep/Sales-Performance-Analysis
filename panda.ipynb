```python
import pandas as pd

# Load CSV file
df = pd.read_csv(r"C:\Users\Administrator\OneDrive\Attachments\Amazon_Data.csv")

# Check rows and columns
print("Rows and Columns:", df.shape)

# First 5 rows
print(df.head())

```

```python
import pandas as pd

# Ensure numeric
df['discounted_price'] = pd.to_numeric(df['discounted_price'], errors='coerce')
df['rating_count'] = pd.to_numeric(df['rating_count'], errors='coerce')

# Calculate Estimated Sales
df['Estimated_Sales'] = df['discounted_price'] * df['rating_count']

# Check first 5 rows
print(df[['product_name', 'discounted_price', 'rating_count', 'Estimated_Sales']].head())
```

```python
import pandas as pd

# Ensure numeric
df['discounted_price'] = pd.to_numeric(df['discounted_price'], errors='coerce')
df['rating_count'] = pd.to_numeric(df['rating_count'], errors='coerce')

# Calculate Estimated Sales
df['Estimated_Sales'] = df['discounted_price'] * df['rating_count']

# Check first 5 rows
print(df[['product_name', 'discounted_price', 'rating_count', 'Estimated_Sales']].head())
```

```python
# Top 5 products
top5_products = df[['product_name', 'Estimated_Sales']].sort_values(by='Estimated_Sales', ascending=False).head(5)

print("Top 5 Best-Selling Products:")
print(top5_products)
```

```python
print(df['discounted_price'].head(10))
print(df['rating_count'].head(10))
```

```python
# Remove ₹ and commas from actual_price
df['actual_price'] = df['actual_price'].replace('[\₹,]', '', regex=True)
df['actual_price'] = pd.to_numeric(df['actual_price'], errors='coerce')

# Create Estimated_Sales (1 unit per product)
df['Estimated_Sales'] = df['actual_price'] * 1

# Check first 5 rows
print(df[['product_name', 'actual_price', 'Estimated_Sales']].head())
```

```python
# Top 5 products by Estimated Sales
top5_products = df[['product_name', 'Estimated_Sales']].sort_values(by='Estimated_Sales', ascending=False).head(5)

print("Top 5 Best-Selling Products:")
print(top5_products)
```

```python
# Show all column names
print(df.columns)
```

```python
# Rank products within each category
df['Category_Rank'] = df.groupby('category')['Estimated_Sales'].rank(ascending=False, method='dense')

# Check first 10 rows
print(df[['product_name', 'category', 'Estimated_Sales', 'Category_Rank']].head(10))
```

```python
# Total spending per customer
customer_sales = df.groupby('user_id')['Estimated_Sales'].sum().reset_index()

# Segment customers
def segment_customer(x):
    if x < 1000:
        return 'Low'
    elif x < 5000:
        return 'Medium'
    else:
        return 'High'

customer_sales['Segment'] = customer_sales['Estimated_Sales'].apply(segment_customer)

# Check first 10 customers
print(customer_sales.head(10))
```

```python
# Split user_id if multiple per row and calculate Estimated_Sales per customer
import pandas as pd

# Explode user_id to have one user per row
df_users = df.copy()
df_users['user_id'] = df_users['user_id'].str.split(',')
df_users = df_users.explode('user_id')

# Group by user_id and sum Estimated_Sales
top_customers = df_users.groupby('user_id')['Estimated_Sales'].sum().sort_values(ascending=False).head(10)

print("Top 10 Customers by Lifetime Value:")
print(top_customers)
```

```python
# Top 10 customers by Estimated Sales
top10_customers = df[['user_id', 'Estimated_Sales']].sort_values(by='Estimated_Sales', ascending=False).head(10)

print("Top 10 Customers by Lifetime Value:")
print(top10_customers)
```

```python
# Rank products within each category by Estimated_Sales
df['Category_Rank'] = df.groupby('category')['Estimated_Sales'].rank(method='max', ascending=False)

# Show top 5 products per category
top_products_per_category = df[df['Category_Rank'] <= 5].sort_values(['category', 'Category_Rank'])
print(top_products_per_category[['category', 'product_name', 'Estimated_Sales', 'Category_Rank']])
```

```python
# Step: Category-wise total sales
category_sales = df.groupby('category')['Estimated_Sales'].sum().sort_values(ascending=False)

# Export CSV for Power BI
category_sales_df = category_sales.reset_index()
category_sales_df.to_csv("category_sales.csv", index=False)

print("category_sales.csv ready for Power BI!")
```

```python
# Customer Segmentation CSV
customer_segmentation = df.groupby('user_id')['Estimated_Sales'].sum().reset_index()

# Segment based on spending
def segment(x):
    if x < 2000:
        return 'Low'
    elif x < 5000:
        return 'Medium'
    else:
        return 'High'

customer_segmentation['Segment'] = customer_segmentation['Estimated_Sales'].apply(segment)

# Export for Power BI
customer_segmentation.to_csv("customer_segmentation.csv", index=False)
print("customer_segmentation.csv ready for Power BI!")
```

```python
# Top Products per Category
df['Category_Rank'] = df.groupby('category')['Estimated_Sales'].rank(method='first', ascending=False)

top_products = df[df['Category_Rank'] == 1]  # only top product per category

# Export CSV
top_products.to_csv("top_products_per_category.csv", index=False)
print("top_products_per_category.csv ready for Power BI!")
```

```python
# Top 10 Customers by Lifetime Value
top_customers = df.groupby('user_id')['Estimated_Sales'].sum().sort_values(ascending=False).head(10).reset_index()

# Export for Power BI
top_customers.to_csv("top_10_customers.csv", index=False)
print("top_10_customers.csv ready for Power BI!")
```
